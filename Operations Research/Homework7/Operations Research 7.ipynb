```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0, 1.5, 10000)
x2 = np.sqrt(1-x1*x1)
z = x1 + x2
x2z = np.sqrt(2) - x1

plt.figure(figsize=(6, 6))
plt.plot(x1, x2, label='x1^2 + x2^2 = 1', color='black')
plt.plot(x1, x2z, label='Z = 2^0.5 = x1 + x2', color='green')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Graph of the nonlinear program')
plt.xlim(0, 1.5)
plt.ylim(0, 1.5)

plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0,10, 10000)
f = 24 * x1 - x1*x1

plt.figure(figsize=(6, 6))
plt.plot(x1, f, label='f(x1)=24x1-x^1', color='black')

plt.xlabel('x1')
plt.ylabel('f(x1)')
plt.title('Objective function over the feasible region of the subproblem 1')
plt.xlim(0, 10)
plt.ylim(0, 150)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0,15, 10000)
f = 10 * x1 - x1*x1

plt.figure(figsize=(6, 6))
plt.plot(x1, f, label='f(x2)=10x2-x^2', color='black')

plt.xlabel('x2')
plt.ylabel('f(x2)')
plt.title('Objective function over the feasible region of the subproblem 2')
plt.xlim(0, 15)
plt.ylim(0, 26)
plt.grid(True)
plt.legend()
plt.show()
```

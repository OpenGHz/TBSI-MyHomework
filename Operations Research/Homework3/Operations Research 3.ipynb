```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0, 10, 1000)
y1 = 3 * x1 + 1
y2 = -2 * x1 + 10
y3 = 4 * x1 - 10
y4 = 0.5 * x1 + 2.5

yz1 = -2/3 * x1 + 6

plt.figure(figsize=(8, 6))
plt.plot(x1, y1, label='-3x1 + x2 = 1', color='green')
plt.plot(x1, y2, label='4x1 + 2x2 = 20', color='blue')
plt.plot(x1, y3, label='4x1 - x2 = 10', color='orange')
plt.plot(x1, y4, label='-x1 + 2x2 = 5', color='purple')

# CPF
a=np.mat('-3,1;-1,2')
b=np.mat('1;5')
cpf1=np.linalg.solve(a,b)
print(f'First CPF is {cpf1}')
a=np.mat('-1,2;4,2')
b=np.mat('5;20')
cpf2=np.linalg.solve(a,b)
print(f'Second CPF is {cpf2}')
a=np.mat('4,-1;4,2')
b=np.mat('10;20')
cpf3=np.linalg.solve(a,b)
print(f'Third CPF is {cpf3}')
a=np.mat('1,0;-3,1')
b=np.mat('0;1')
cpf4=np.linalg.solve(a,b)
print(f'Fourth CPF is {cpf4}')
a=np.mat('1,0;-3,1')
b=np.mat('0;1')
cpf4=np.linalg.solve(a,b)
print(f'Fourth CPF is {cpf4}')
a=np.mat('0,1;4,-1')
b=np.mat('0;10')
cpf5=np.linalg.solve(a,b)
print(f'Fifth CPF is {cpf5}')
a=np.mat('0,1;1,0')
b=np.mat('0;0')
cpf6=np.linalg.solve(a,b)
print(f'Sixth CPF is {cpf6}')

plt.scatter([cpf1[0],cpf2[0],cpf3[0],cpf4[0],cpf5[0],cpf6[0]], [cpf1[1],cpf2[1],cpf3[1],cpf4[1],cpf5[1],cpf6[1]],
            marker='o', edgecolors='k', facecolors='none',s=100)

plt.plot(x1, yz1, label='Z = 18 = 2x1 + 3x2', color='red')

y_upper = np.piecewise(x1, [x1 < 0.6, (x1 >= 0.6) & (x1 < 3), (x1>=3) & (x1<10/3)], [lambda x: 1 + 3*x, lambda x: 2.5 + 0.5*x, lambda x: 10 -2*x])
y_down = np.piecewise(x1, [x1 < 2.5, (x1 >= 2.5) & (x1<10/3)], [lambda x: 0, lambda x: -10 + 4*x])

plt.fill_between(x1, y_upper, y_down,
                color='gray', alpha=0.5,
                label='Feasible Region')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Graphical solution of the linear program')
plt.xlim(-0.2, 10)
plt.ylim(-0.2, 10)

plt.grid(True)
plt.legend()
plt.show()
```
